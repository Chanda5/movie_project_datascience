```python
#Classify movies based on ratings(Excelent,Good,Average)
data.columns
```

```python
def rating(rating):
    if rating>=8.0:
        return 'Excellent'
    elif rating>=7.0:
        return 'Good'
    else:
        return 'Average'
```

```python
data['rating_category']=data['Rating'].apply(rating)
data.head()
```
